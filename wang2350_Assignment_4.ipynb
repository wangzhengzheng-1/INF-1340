Q1. Print average shipping cost for items in different departments, which department has the most and least expensive shipping cost?

```python
ave_cost=superstore.groupby(['Department'])['Shipping Cost'].mean()
print(ave_cost)
```

```python
print('The Furniture has the most expensive shipping cost', max(ave_cost))
print('The Office Supplies has the least expensive shipping cost',min(ave_cost))
```

Q3. Using line chart, compare the total sale among three most profitable states throughout the 4 years of available data. 

```python
#The three most profitable states

superstore.groupby(['State'])['Profit'].sum().sort_values(ascending=False)[0:3]
```

```python
superstore1=superstore[superstore['State']=='Illinois']
superstore2=superstore[superstore['State']=='New York']
superstore3=superstore[superstore['State']== 'Texas'  ]
superprofit=pd.concat([superstore1,superstore2,superstore3],axis=0)
superprofit
```

```python
superprofit_=superprofit.groupby(['Order Date','State']).Sales.sum()
superprofit_
```

```python
superprofit_.unstack(fill_value=0).plot()
plt.ylabel('The total sales')
plt.show()
```

Q4. Using bar chart, compare average shipping cost for different kinds of shipping and different kinds of containers. How much is the average shipping cost for small box containers in express air shipments?

```python
avecost=superstore.groupby(['Ship Mode','Container'])['Shipping Cost'].mean().unstack().reset_index().plot.bar(x='Ship Mode')
plt.ylabel('Average Shipping Cost')
plt.show()
```

```python
ave_small_air=superstore['Shipping Cost'][(superstore['Container']=='Small Box') & (superstore['Ship Mode']=='Express Air')].mean()
print('The average shipping cost for small box containers in express air shipments is ',ave_small_air)
```
